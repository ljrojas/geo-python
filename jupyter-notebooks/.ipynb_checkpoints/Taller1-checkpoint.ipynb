# 1. Configuración de Ambiente

## ¿Qué es un ambiente? 

Un ambiente (o enterno virtual) es un espacio de trabajo que contiene las herramientas y librerías necesarias para un proyecto o un conjunto de proyectos.

Este espacio de trabajo se puede pensar como el estudio de un artista, en el cual trabaja distintos lienzos. Cada lienzo, por su composición, podría requerir pinturas especiales, solo compatibles con su material. El artista podría tener un una habitación los materiales acrílicos, y en otra los óleos, ya que si se mezclaran los materiales podría dañar el lienzo.

Con este mal ejemplo, buscamos ilustrar que cada proyecto que trabajemos tendrá características propias. Así como cada librería que usemos. Las librerías presentan interdependencias, y además tienen distintas versiones en el tiempo, las cuales pueden dejar obsoletas ciertas funciones.

La idea del ambiente entonces, es aislar estos espacios de trabajo para que sigan funcionando a su propio ritmo, sin preocuparnos de las distintas versiones que podamos estar usando en cada uno. 

Anaconda es un herramienta que nos permite gestionar estos ambientes, con un enfoque hacia la ciencia de datos. 

![title](../docs/conda-env2.jpeg)

## Trabajando con ambientes

Para trabajar con ambientes usaremos la línea de comandos. En Windows Anaconda instaló **Anaconda Prompt**, en linux o mac se usa la terminal corriente.

Se recomienda tener a mano el resumen de comandos básicos (a.k.a *cheat sheet* o *torpedo*), el cual puede ser encontrado aquí: https://docs.conda.io/projects/conda/en/4.6.0/_downloads/52a95608c49671267e40c689e0bc00ca/conda-cheatsheet.pdf

### Información de conda

Una vez abierta Anaconda Prompt, podemos obtener información de conda con el siguiente comando `conda info`. 

Además, para explorar en más detalle podemos tipear `conda help`.

### Listado de ambientes 

Para tener un listado de los ambientes dentro del sistema, tipeamos `conda env list`. Si no hemos creado ninguno, solo debiese aparecer el entorno base, que contine la mayoría de las librerías necesarias para la ciencia de datos. El entorno activo estará destacado con un asterisco en la lista, además de aparecer entre parentesis al inicio del prompt.

###  Crear un ambiente 

Ahora para crear nuestro nuevo ambiente tipeamos `conda create --name nombre_nuevo_ambiente`. El sistema solicitará confirmación para realizar la operación. Recomendamos usar un nombre no complejo que sea representativo del tema a trabajar, como por ejemplo *geo-taller*.

### Activar el nuevo ambiente

Podemos verificar que nuestro ambiente fue creado tipeando `conda env list`. Si este aparece en la lista, lo activamentos con el comando `conda activate nombre_nuevo_ambiente`. Así mismo, para desactivarlo usamos el comando `conda deactivate`.

### Listado de paquetes y librerías del ambiente

Una vez activado el nuevo ambiente, podemos verificar los paquetes que contiene con el comando `conda list`. Este comando debería entregar una lista vacía, ya que no hemos instalado ninguna librería.

### Eliminar un ambiente

Si deseasemos eliminar algun entorno, primero checkeamos que no esté activado. De estar activado, salemos de él con `conda deactivate`, luego lo eliminamos con 

`conda env remove --name nombre_ambiente`.

### 4.3 Datos espaciales: Sistemas de Referencia de Coordenadas

Al operar con distintas capas de información, de distinta fuente, debemos ser cuidadosos en usar el mismo sistema de referencia en nuestras operaciones espaciales ('*comparar peras con peras*'). 

En el caso de los puntos de salud, fueron cargados de una tabla excel. Sabemos que esta representa latitud y longitud, y en nuestro objeto espacial debiese ser así. 

Ahora revisamos en qué sistema de referencia se encuentra. Esto lo hacemos accediendo al atributo crs de nuestra GeoSerie.

```python
print(salud.crs) 
```

Vemos que no tiene ningún sistema asignado, lo que tiene sentido dado que cargamos los datos a partir de una tabla.

Ya que sabemos que la información de coordenadas se encuentra en latitud y longitud, asignaremos esta sistema de referencia al objeto.

Mayor información respecto del manejo de proyecciones y sistemas de referencia en geopandas puede encontrarse en la documentación en línea: https://geopandas.org/projections.html

Cabe destacatar que definir el sistema de referencia de los datos trabajados **no es lo mismo que reproyectar**.

Una rápida revisión nos muestra que la proyección que buscamos es WGS84 Latitude/Longitude: "EPSG:4326"

```python
salud.crs = 'EPSG:4326' # con esto seteamos la proyeccion a la que conocemos
salud.crs        # esto NO ES LO MISMO QUE RE PROYECTAR
```

Manejar correctamente las proyecciones es un aspecto clave de revisar al trabajar con datos espaciales. Una vez que ya tiene asignada la proyección, podemos re-proyectarla (pasar de grados a metros con una proyección local por ejemplo). Algunas operaciones como calcular el area requerirá que la información esté en ciertas unidades de medida.

Ahora revisamos las proyecciones de la división comunal:

```python
# revisar CRS
comunas_novena.crs
```

Esta proyección es distinta de la asignada recientemente, sin embargo también usa grado de latitud y longitud. Podemos encontrar más información en: https://epsg.io/4674 

```python
comunas_novena = comunas_novena.to_crs("EPSG:4326")
comunas_novena.head()
```

Ahora retomamos los datos de DEIS y procedemos a darles el contexto que necesitan. 

Para esto definimos un mapa de base, las comunas, y luego graficamos la capa de puntos de salud encima, diferenciando cada tipo de centro de salud con un color distinto. 

Para facilitar la lectura incluimos también la leyenda.

```python
# Definir mapa base
base = comunas_novena.plot(color = 'white', edgecolor = 'black', figsize = (12,12))

# Añadimos la capa de puntos indicando el mapa base
salud.plot(ax = base, marker = '*', cmap = 'jet', markersize = 10, column = 'tipo', legend = True)

# rescatar la leyenda y reubicar
leg = base.get_legend()
leg.set_bbox_to_anchor((1, 1.))
leg.set_title('Tipo de Establecimiento')
# Definimos titulo
base.set_title('Centros de Salud en la Novena Región', fontsize=20)
base.set_axis_off()

# por último podemos añadir los nombres de las comunas mediante la siguiente operacion
comunas_novena.apply(lambda x: base.annotate(s=x.DESC_COMUN, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
```

Sin embargo esta presentación tampoco entrega tanta información.

A continuación intentaremos usar un *mapa de calor* para intentar reflejar la concentración de puntos de atención en el espacio. Para eso usamos la librería `geoplot`

```python
import geoplot

# mapa base
ax = geoplot.kdeplot(
    salud,
    clip = comunas_novena.geometry,
    shade=True, 
    cmap='Greens',
    figsize = (12,12))

# titulo
ax.set_title('Concentración de Centros de Salud en la Novena Región', fontsize=20)

# grafica el mapa
geoplot.polyplot(comunas_novena, ax=ax, zorder=1)

# añade capa de puntos
geoplot.pointplot(salud, ax = ax)

# añadimos nombres de las comunas
comunas_novena.apply(lambda x: ax.annotate(s=x.DESC_COMUN, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
```

### 5 Datos Espaciales: Operaciones Básicas

Para ilustrar el uso de algunas operaciones espaciales usaremos datos de cuerpos de agua y manzanas demográficas. 

Estos pueden ser encontrados en:

- Cuerpos de agua:  https://ide.mma.gob.cl/
- Manzanas: http://geoine-ine-chile.opendata.arcgis.com/datasets/54e0c40680054efaabeb9d53b09e1e7a_0

La idea es calcular cuantas personas viven dentro de un radio de los cuerpos de agua principales de la región.

Para esto usaremos las siguientes operaciones:

1. Calcular el Area de un polígono
2. Transformar polígonos a puntos  (centroide)
3. Filtrar espacialmente
4. Buffer
5. Spatial join

Primero calcularemos el Area de los cuerpos de agua de la región para quedarnos con los más grandes. 

Luego generaremos un buffer de 1 kilometro para seleccionar espacialmente los puntos de población relevantes. 

Primero, seleccionaremos los cuerpos de agua que estén dentro de la región, usando sus centroides.

```python
# RECORDAR CAMBIAR LA RUTA A LA CARPETA MAESTRA DONDE DESCARGARON LOS DATOS
cuerpos_de_agua = gpd.read_file('D:/data/CuerposdeAgua/CuerposdeAgua.shp')
```

Primero verificamos la proyección en que se encuentran los cuerpos de agua.

```python
cuerpos_de_agua.crs
```

Se aprecia que esta proyección tiene metros como unidad de medida. 

Esto nos servirá para calcular el area y para generar los buffers más adelante.

```python
# seleccionamos la información de interés
cuerpos_de_agua = cuerpos_de_agua.loc[:,['OBJECTID','Nombre','geometry']]
cuerpos_de_agua.head()
```

Aprovechamos de ilustrar la diferencia de resultados de la misma operación al cambiar la proyección en la que se encuentran los datos.

```python
# Ilustramos la diferencia entre proyección al calcular el area del primer poligono de la muestra
print('Area usando utm: ' + str(cuerpos_de_agua.loc[0,'geometry'].area))
print('Area usando lat lon: ' + str(cuerpos_de_agua.to_crs('EPSG:4326').loc[0,'geometry'].area))
```

Procedemos a seleccionar espacialmente los datos de cuerpos de agua que se encuentran en el area de estudio.

Para esto re-proyectamos y llevamos a centroide (el punto del centro), ya que es menos intensivo computacionalmente verificar si un punto cae dentro de un polígono.

```python
# Re proyectamos
cda_pto = cuerpos_de_agua.to_crs('EPSG:4326')
# Re define la columna geometry que contiene la información espacial.
cda_pto['geometry'] = cda_pto['geometry'].centroid
# inspección
cda_pto.head(10)
```

Para chequear qué puntos caen dentro de la región, importamos shapely, la librería que esta en la base de geopandas para manejar los datos espaciales. 

Con esto haremos la consulta sobre qué puntos caen dentro de la región. Esto con el objetivo de construir una mascara booleana para seleccionar los polígonos relevantes.

```python
import shapely
import shapely.speedups

shapely.speedups.enable()

# re proyectamos
r = novena_region.to_crs('EPSG:4326')
p = cda_pto.geometry

p.within(r).value_counts()
```

La operación debiese entregar un array de valores booleanos, que usaremos para seleccionar los datos relevantes para el estudio. Sin embargo, según esto, ningún punto cae dentro de la región. Esto no tiene sentido. 

Revisamos en detalle la región:

```python
r
```

Vemos que la geometría es un MULTIPOLÍGONO, es decir, una lista de distintos polígonos. Esto no salta a la vista facilmente al graficar la región. Puede que efectivamente no esté considerando el interior de la superficie al momento de filtrar los puntos. 

Usaremos la operación espacial `unary_union` que retorna una geometría que contiene la union de las geometrías dentro del multipolígono.

```python
r.unary_union
```

Repetimos la operación con esta nueva geometría.

```python
p.within(r.unary_union).value_counts()
```

¡Éxito! Ahora usaremos esta máscara de valores lógicos para filtrar la información de cuerpos de agua.

```python
mask = p.within(r.unary_union)
```

```python
cuerpos_novena = cuerpos_de_agua.loc[mask]
cuerpos_novena.head()
```

Ya que vemos que los puntos están en utm, procedemos a generar una nueva columna de Area. 

```python
cuerpos_novena.insert(2,'Area',cuerpos_novena.area / 1000000) # llevamos a kilometros cuadrados
cuerpos_novena.head()
```

Inspeccionamos la distribución de la variable calculada:

```python
cuerpos_novena.loc[:,'Area'].describe()
```

Nos quedaremos con el 5% de cuerpos de agua más grandes.

```python
# percentil 95
p95 = np.percentile(cuerpos_novena['Area'],95)

# mascara de seleccion
mask = cuerpos_novena['Area'] >= p95

cuerpos_mayores = cuerpos_novena.loc[mask].sort_values(by='Area',ascending=False)
cuerpos_mayores
```

```python
# guardamos los resultados
cuerpos_mayores = cuerpos_novena.loc[mask]
cuerpos_mayores.to_file('../data/cuerpos_mayores.shp')
```

Inspección visual:

```python
# Definir mapa base
base = comunas_novena.plot(color = 'white', edgecolor = 'black', figsize = (12,12))

# Añadimos la capa de puntos indicando el mapa base
cuerpos_mayores.to_crs('EPSG:4326').plot(ax = base)
# Definimos titulo
base.set_title('Principales cuerpos de Agua de la Novena Región', fontsize=20)
base.set_axis_off()

# por último podemos añadir los nombres de las comunas mediante la siguiente operacion
comunas_novena.apply(lambda x: base.annotate(s=x.DESC_COMUN, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
```

Ahora cargamos las manzanas de población urbana de la región.

```python
poblacion = gpd.read_file('../data/manzanas_novena.shp')
```

Revisamos en qué proyección encontramos el dato:

```python
poblacion.crs
```

Ahora re-proyectamos y pasamos a punto.

```python
# re proyectamos
poblacion = poblacion.to_crs('EPSG:4326')
# transformamos a punto
poblacion['geometry'] = poblacion['geometry'].centroid
```

Con esto, podemos darle un buffer a los cuerpos de agua y verificar cuánta gente vive dentro de este buffer.

```python
# bufer cuerpos de agua
area_interes = cuerpos_mayores.copy()
area_interes['geometry'] = cuerpos_mayores.to_crs('EPSG:32719').buffer(5000)
area_interes.plot()
```

Inspección visual de los puntos de población:

```python
poblacion.plot()
```

A partir de este gráfico podemos apreciar que los puntos de población no son completos, probablemente dejan fuera los puntos de población rural (recordar distribución de puntos de centros de asistencia).

Ahora usaremos un join espacial para asignarle la información de una capa (población) a otra (cuerpos de agua con buffer)

```python
# sjoin
join = gpd.sjoin(area_interes.to_crs('EPSG:4326'), poblacion, op = 'contains', how = 'left')
join
```

Se Aprecia que hay valores missing, debido a que no contamos con la población rural.

```python
join[['Nombre','TOTAL_PERS']].fillna(0).groupby('Nombre').agg('sum')
```

Con esto tenemos la cantidad de población urbana que vive en un radio de 5 kilometros de los cuerpos de agua principales de la región.
