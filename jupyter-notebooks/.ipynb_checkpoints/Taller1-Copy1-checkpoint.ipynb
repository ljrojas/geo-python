## Pandas: Carga de datos
Para iniciar el camino hacia los datos espaciales, usamos las dataframe que nos entrega la librería pandas. 
Como veremos, algunas de sus funciones serán claves en el correcto ensamblaje de el tipo de dato espacial.

En este taller, usaremos la base de datos de MINSAL de establecimientos de salud, disponible en: https://repositoriodeis.minsal.cl/DatosAbiertos/Establecimientos_ChileDEIS_MINSAL(05-06-2020).xlsx

Usamos la librería os (operational system) para ayudarnos a ubicar los archivos y carpetas en donde está la información descargada.

```python
os.getcwd() # en que directorio nos encontramos
os.listdir('../data') # .. sube un directorio en el arbol de carpetas

# En windows podemos usar el tipo de dato raw string para guardar rutas con backslash ("\")
# r'C:\Users\Downloads\data'

file = '../data/Establecimientos_ChileDEIS_MINSAL(05-06-2020).xlsx'
deis = pd.read_excel(io=file, header=1, decimal='.')

# Inspección
deis.dtypes.head()
deis.dtypes.tail()
```

Se aprecia al final de la tabla la información básica para definir el dato espacial más simple: el punto. 
Para proyectar correctamente un punto, necesitaremos además el sistema de referencia. En este caso, sabemos que son grados decimales.

## Pandas: Limpieza de datos

Para facilitarnos el acceso a la información procedemos a:
- seleccionar variables
- renombrar variables

```python
# listado de variables
deis.columns
```

Mediante un simple copy paste seleccionamos las variables de interés, luego usaremos un diccionario para renombrarlas.

```python
# Seleccionar columnas de interés
deis_cols = deis.filter(['Código Región','Región','Código Comuna',
                         'Nombre Comuna','Tipo Establecimiento',
                         'LONGITUD [Grados decimales]','LATITUD      [Grados decimales]'])

# renombrar columnas
dic_cols = {'Código Región':'cod_reg',
            'Región':'nom_reg',
            'Código Comuna':'cod_com',
            'Nombre Comuna':'nom_com',
            'Tipo Establecimiento':'tipo',
            'LONGITUD [Grados decimales]':'lon','LATITUD      [Grados decimales]':'lat'}

deis_cols = deis_cols.rename(columns = dic_cols )

# inspección
deis_cols.head()
```

```python
# Revision base de datos

# - numero de establecimientos por region
deis_cols.groupby('cod_reg').agg({'cod_com':['count']}).transpose()
# deis_cols.loc['cod_reg'].values()
```

```python
import numpy as np
df = deis_cols
table = pd.pivot_table(df, index=['tipo'],
                    columns=['cod_reg'], aggfunc='count')
table.fillna('-')

import seaborn as sns
import matplotlib
cm = sns.light_palette("green", as_cmap=True)


table.fillna('-')
```

```python
from IPython.html import widgets
@widgets.interact
def magnify():
    return [dict(selector="th",
                 props=[("font-size", "4pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "12pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '200px'),
                        ('font-size', '12pt')])
]


cmap = cmap=sns.diverging_palette(5, 250, as_cmap=True)


table.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '1pt'})\
    .set_caption("Hover to magnify")\
    .set_precision(2)\
    .set_table_styles(magnify())
```

```python
# tipos de establecimientos en la novena region

deis_novena = deis_cols.loc[deis_cols['cod_reg']==9]
deis_novena.groupby(['tipo']).agg('count')

deis_novena.filter(['cod_reg','tipo']).groupby('tipo').agg('count').sort_values(by='cod_reg').plot.barh()

# incluir titulo
```

```python
# Espacializar
# checkear tipos de datos
deis_novena.dtypes
```

```python
# filtrar entradas sin información espacial

# donde encontramos mayor información disponbble
deis_novena.loc[deis_novena['lat']!='Pendiente'].groupby(['nom_com']).agg('count').sort_values(by='cod_reg',ascending=False).head(10)
```

```python
# Espacializar informacion
df = deis_novena[deis_novena['lat']!='Pendiente']
df[['lat','lon']] = df[['lat','lon']].astype('float')
```

```python
salud = gpd.GeoDataFrame(df,
                geometry = gpd.points_from_xy(df.lon, df.lat))
```

```python
import matplotlib 
salud.plot()
```
